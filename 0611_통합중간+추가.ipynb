2. 평균 제곱 오차(MSE)는 mean_squared_error 함수를 사용하여 실제 레이블과 예측값의 차이의 제곱을 평균한 값이다. MSE는 회귀 문제에서 예측값과 실제 값 사이의 오차를 측정하는 데 사용된다. MSE가 낮을수록 모델의 예측이 더 정확하다.

3. 결정 계수(R^2)는 r2_score 함수를 사용하여 계산된다. R^2는 회귀 모델의 예측이 종속 변수의 분산을 얼마나 잘 설명하는지를 나타내는 지표이다. R^2는 0부터 1까지의 값을 가지며, 1에 가까울수록 모델의 예측이 더 잘 설명한다.

4. 평균 절대 오차 (MAE, Mean Absolute Error): 실제 값과 예측 값의 차이의 절대값을 평균한 값이다. MAE는 오차의 크기를 나타내며, 작을수록 모델의 예측이 더 정확하다.

5. 평균 제곱근 오차 (RMSE, Root Mean Squared Error): 실제 값과 예측 값의 차이의 제곱을 평균한 뒤 제곱근을 취한 값이다. RMSE는 MSE의 제곱근으로, 오차의 크기를 나타내며, 작을수록 모델의 예측이 더 정확하다.

# 1. 로지스틱회귀

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score
from sklearn.metrics import mean_absolute_error
```

```python
# 예측에 사용할 특성 열 선택
features = ['자치구', '승하차', '노인인구', '총노인시설', '노인시설정원','노인인프라구축현황']

# 원-핫 인코딩 수행
df_encoded = pd.get_dummies(df, columns=['자치구'])

# 예측에 사용할 특성 열 선택
features_encoded = df_encoded.columns.drop(['레이블'])

# 훈련 데이터와 테스트 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(df_encoded[features_encoded], df_encoded['레이블'], test_size=0.2, random_state=42)

# 로지스틱 회귀 모델 구축
model = LogisticRegression()
model.fit(X_train, y_train)

# 예측 수행
y_pred = model.predict(X_test)

# 정확도 평가
accuracy = accuracy_score(y_test, y_pred)
print('정확도:', accuracy)

# 모델 평가
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("평균 제곱 오차(MSE): ", mse)
print("결정 계수(R^2): ", r2)
print("평균 절대 오차(MAE): ", mae)
print("평균 제곱근 오차(RMSE): ", rmse)
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score

# f1-score 계산
f1 = f1_score(y_test, y_pred)
print("f1-score:", f1)

# 정밀도 계산
precision = precision_score(y_test, y_pred)
print("정밀도:", precision)

# 재현율 계산
recall = recall_score(y_test, y_pred)
print("재현율:", recall)
```

# 4. AdaBoost

```python
import pandas as pd
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# '자치구' 열을 기준으로 원-핫 인코딩수행.
df_encoded = pd.get_dummies(df, columns=['자치구'])

# 예측에 사용할 특성 열 선택, '레이블' 열을 제외한 모든 열이 특성 열로 선택
features_encoded = df_encoded.columns.drop(['레이블'])

# training 데이터와 test 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(df_encoded[features_encoded], df_encoded['레이블'], test_size=0.2, random_state=42)

# AdaBoost classifier 생성
ada_boost = AdaBoostClassifier()
ada_boost.fit(X_train, y_train)

# 테스트 데이터 예측 생성
y_pred = ada_boost.predict(X_test)

# 모델 평가
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("평균 제곱 오차(MSE): ", mse)
print("결정 계수(R^2): ", r2)
print("평균 절대 오차(MAE): ", mae)
print("평균 제곱근 오차(RMSE): ", rmse)
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score

# f1-score 계산
f1 = f1_score(y_test, y_pred)
print("f1-score:", f1)

# 정밀도 계산
precision = precision_score(y_test, y_pred)
print("정밀도:", precision)

# 재현율 계산
recall = recall_score(y_test, y_pred)
print("재현율:", recall)
```

- AdaBoost: 약한 모델에서 잘못 예측한 샘플에 가중치를 높여서 다음 모델이 그 샘플에 더 집중하도록 하는 것. 즉, 이전 모델의 오류를 보완하도록 다음 모델을 학습시킴. 이러한 반복적인 학습과 가중치 갱신을 통해 전체 앙상블 모델은 점점 더 강력한 예측 능력을 갖게 됨.
- 정확도: 모델이 정확하게 예측한 비율을 나타내는 지표. 0부터 1까지의 범위에서 표현되며, 1에 가까울수록 모델의 예측이 정확함.
- 위의 코드에서 ada_boost.predict(X_test)를 통해 테스트 세트에 대한 예측을 수행하고, 이 예측 결과를 y_pred에 저장. 그 다음, accuracy_score(y_test, y_pred)를 통해 실제 타겟 변수인 y_test와 예측된 결과인 y_pred를 비교하여 정확도를 계산.
- 정확도가 1.0이 출력되었지만, 정확도만으로는 모델의 성능을 완전히 평가할 수는 없음.다른 평가 지표인 정밀도, 재현율, F1 점수 등을 함께 고려하여 모델의 성능을 더욱 종합적으로 평가, 해석 필요.

# 5. LightGBM (민정님 - 내 거에서는 오류)

```python
import lightgbm

# 예측에 사용할 특성 열 선택
features = ['자치구', '승하차', '노인인구', '총노인시설', '노인시설정원','노인인프라구축현황']

# 원-핫 인코딩 수행
df_encoded = pd.get_dummies(df, columns=['자치구'])

# 예측에 사용할 특성 열 선택
features_encoded = df_encoded.columns.drop(['레이블'])

# 훈련 데이터와 테스트 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(df_encoded[features_encoded], df_encoded['레이블'], test_size=0.2, random_state=42)

# LightGBM 데이터셋 생성
train_data = lightgbm.Dataset(X_train, label=y_train)

# LightGBM 모델 설정
params = {
    'objective': 'binary',
    'metric': 'binary_logloss',
    'boosting_type': 'gbdt',
    'num_leaves': 31,
    'learning_rate': 0.05,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': 0
}

# LightGBM 모델 학습
model = lightgbm.train(params, train_data, num_boost_round=100)

# 예측 수행
y_pred = model.predict(X_test)
y_pred_binary = [1 if x >= 0.5 else 0 for x in y_pred]

# 정확도 평가
accuracy = accuracy_score(y_test, y_pred_binary)
print('정확도:', accuracy)

# 모델 평가
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("평균 제곱 오차(MSE): ", mse)
print("결정 계수(R^2): ", r2)
print("평균 절대 오차(MAE): ", mae)
print("평균 제곱근 오차(RMSE): ", rmse)
```

- LightGBM은 경사 부스팅 알고리즘을 기반으로 한 빠른 속도와 높은 성능을 제공하는 머신러닝 모델
- LightGBM은 lightgbm 라이브러리를 통해 사용가능

# +  앙상블 - AdaBoost, XGBoost, LightGBM

1. AdaBoost(Adaptive Boosting): AdaBoost는 이전 모델이 잘못 예측한 샘플에 가중치를 부여하여 다음 모델을 학습시키는 방식으로 약한 학습기(weak learner)를 강한 학습기(strong learner)로 변환, 잘못 분류된 샘플에 집중하여 예측을 개선하는 특징

2. XGBoost(eXtreme Gradient Boosting): XGBoost는 그래디언트 부스팅 알고리즘을 기반으로 한 빠르고 고성능의 앙상블 모델. 트리 기반 학습 알고리즘을 사용하며, 그래디언트 부스팅의 단점인 계산 속도와 과적합을 개선한 알고리즘입니다. XGBoost는 분류와 회귀 문제에 모두 사용될 수 있다

3. LightGBM(Light Gradient Boosting Machine): LightGBM은 XGBoost와 유사한 그래디언트 부스팅 알고리즘을 사용하는 앙상블 모델. 대량의 데이터를 처리하는 데 특화되어 있으며, 빠른 학습 속도와 낮은 메모리 사용량을 가지고 있다. LightGBM은 카테고리형 특성과 결측값 처리를 자동으로 지원하며, 범주형 변수의 원-핫 인코딩을 수행하지 않아도 되는 장점이 있다.
