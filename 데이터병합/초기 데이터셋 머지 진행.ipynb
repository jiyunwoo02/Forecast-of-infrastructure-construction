```python
# 출발 데이터 통합
gu1 = S1['자치구'].value_counts()
gus1 = gu1.sort_values(ascending=False)
gu2 = S2['자치구'].value_counts()
gus2 = gu2.sort_values(ascending=False)
gu3 = S3['자치구'].value_counts()
gus3 = gu3.sort_values(ascending=False)
gu4 = S4['자치구'].value_counts()
gus4 = gu4.sort_values(ascending=False)

GUS = pd.concat([gus1, gus2, gus3, gus4], axis=1)
GUS.head(10)
```

```python
# 출발 데이터 결측치 0으로 대체
GUSS = GUS.fillna(0)
GUSS.head(10)
```

```python
GUSS.reset_index(drop = False, inplace = True)
GUSS.head()
```

```python
# 인덱스 이름 변경 & 다중 인덱스 정렬
GUSS.columns = ['자치구','13시','14시','15시','16시']
GUSS.head(10)
```

```python
care = pd.read_csv('노인복지.csv')
care.head()
```

```python
seoul = pd.merge(GUSS, care, on='자치구')
seoul
```

```python
print(seoul.info())
```

```python
seoul.apply(pd.to_numeric, errors = 'coerce')
seoul['16시'] = pd.to_numeric(seoul['16시'])
seoul['노인인구'] = pd.to_numeric(seoul['노인인구'].str.replace(',',''))
seoul['요양시설정원'] = pd.to_numeric(seoul['요양시설정원'].str.replace(',',''))
```

```python
# encording
from sklearn import preprocessing

def encode_seoul(seoul):
    features = ['자치구']
    for feature in features:
        le = preprocessing.LabelEncoder()
        le = le.fit(seoul[feature])
        seoul[feature] = le.transform(seoul[feature])

    return seoul
```

```python
seoul = encode_seoul(seoul)
print(seoul.info())
# 데이터 셋이 모두 숫자형태로 변경됨
```

```python
import csv
seoul.to_csv("seoul.csv")
```
