## 시청역 승차 시간대별 분포

```python
# 필터링할 역명
station_name = '시청'

# 역명이 시청이고 승차인 데이터 필터링
filtered_data = data[(data['역명'] == station_name) & (data['승하차구분'] == '승차')]

# 시간대 열들
time_columns = ['06시간대이전', '06-07시간대', '07-08시간대', '08-09시간대', '09-10시간대', 
                '10-11시간대', '11-12시간대','12-13시간대','13-14시간대','14-15시간대',
                '15-16시간대','16-17시간대','17-18시간대','18-19시간대','19-20시간대',
                '20-21시간대','21-22시간대','22-23시간대','23-24시간대','24시간대이후']

# 시간대별 총 인원 수 계산
total_passengers = filtered_data[time_columns].sum()

# 그래프 설정
plt.figure(figsize=(10, 6),dpi=200)
total_passengers.plot(kind='bar', color='blue')
plt.xlabel('시간대')
plt.ylabel('총 인원 수')
plt.title(f'{station_name}역 시간대별 총 승차인원')
plt.show()

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'
```

- 시청역에서는 18-19시간대에 승차인원 분포가 가장 높다!
- station_name에 다른 역들 넣으면 분포 확인가능
